```python
import numpy as np
import skimage.filters
from matplotlib import pyplot as plt
from FM_Method_3c import *
from Potencial3D import *
from time import time
```

```python
def proceso( image , G , save , sigma = 2.5 , b = 0.7 , cut = 300 ):  
    start = time()
    Dimage = skimage.filters.gaussian(image, 
            sigma=(sigma, sigma,sigma), truncate=3.5, multichannel=True)  
    PDimage = potencial3D( Dimage )
    Fimage = np.exp( -b*PDimage)
    Timage = solve_Eikonal( Fimage , Gamma = G , Delta = 1 , cutOff = cut )
    end = time()
    print( "tiempo = ", end - start)
    #save : Ruta de guardado
    np.save( save , Timage )
```
